```python
#실습 1번

import csv
import matplotlib.pyplot as plt

f = open('seoul_10.csv', 'r', encoding='cp949') #2011.05.26 ~ 2021.05.26
data = csv.reader(f)
header = next(data)
temp_low = []
temp_high = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '26':
            temp_high.append(float(row[-1]))
    if row[-2] != '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '26':
            temp_low.append(float(row[-2]))
f.close()
plt.plot(temp_high, color='magenta')
plt.plot(temp_low, color='cyan')

plt.show()
```

```python
#실습 2번

import csv
import matplotlib.pyplot as plt

f = open('seoul_10.csv', 'r', encoding='cp949') #2011.05.26 ~ 2021.05.26
data = csv.reader(f)
header = next(data)
temp_1 = []
temp_8 = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '01':
            temp_1.append(float(row[-1]))
        if row[0].split('-')[1] == '08':
            temp_8.append(float(row[-1]))
f.close()

plt.hist(temp_8, bins=100, color='r', label='Aug')
plt.hist(temp_1, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```
